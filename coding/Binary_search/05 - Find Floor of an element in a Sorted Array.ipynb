 

# Find Floor of an Element in a Sorted Array Handout

This handout explains how to determine the floor of a given target value in a sorted array. The **floor** of a number *x* in an array is defined as the greatest element in the array that is less than or equal to *x*. We achieve this efficiently using a modified binary search algorithm.

---

## 1. IP–OP–PS (Input, Output, Problem Statement)

### Problem Statement
Given a **sorted array** and a target value *x*, find the floor of *x* in the array. The floor is the largest element that is less than or equal to *x*. If no such element exists (i.e., if *x* is smaller than the smallest element), the function should indicate that a floor does not exist.

**Key Operations**:
- **Search** the array using binary search.
- **Maintain** a candidate floor by updating it when an element ≤ *x* is found.
- **Return** the candidate floor or indicate that none exists.

### Inputs
- **Array**: A sorted list of integers.
- **Target (x)**: The value for which we need to find the floor.

### Outputs
- The **floor value** (the greatest element in the array that is ≤ *x*), or an indicator (such as `-1`) if no such element exists.

### Detailed Example

**Sample Input**:
```plaintext
Array: [1, 2, 8, 10, 10, 12, 19]
Target (x): 5
```

**Sample Output**:
```plaintext
Floor: 2
```

*Explanation*:  
Among the elements less than or equal to 5, the largest element is 2.

---

## 2. Identification

### Why Use Modified Binary Search?

1. **Sorted Array Advantage**:
   - Since the array is sorted, binary search can be applied to quickly narrow down the candidate floor.
   
2. **Efficiency**:
   - A simple linear scan takes O(n) time, but binary search finds the floor in O(log n) time by halving the search space at each iteration.

3. **Key Cues**:
   - The sorted property of the array and the need to locate an element based on a comparison (≤ *x*) point directly to an adapted binary search strategy.

---

## 3. Break Down → Modified Binary Search Approach

### Step-by-Step Sub-Tasks

1. **Initialization**:
   - Set two pointers: `low = 0` and `high = n - 1` (where `n` is the length of the array).
   - Initialize a variable `floorIndex = -1` to keep track of the best candidate (starting with none).

2. **Iterative Process**:
   - **Compute Midpoint**:  
     Calculate `mid = low + (high - low) / 2`.
   - **Compare with Target**:
     - If `array[mid]` is equal to *x*, then *x* is the floor (as it is exactly present). Return `array[mid]`.
     - If `array[mid]` is less than *x*, update `floorIndex` to `mid` (as this is a candidate) and move the `low` pointer to `mid + 1` to search for a larger candidate.
     - If `array[mid]` is greater than *x*, move the `high` pointer to `mid - 1` to search the left part of the array.

3. **Termination**:
   - When `low` exceeds `high`, the search is complete.
   - If a candidate was found (`floorIndex` ≠ -1), return the element at that index as the floor.
   - Otherwise, indicate that no floor exists.

4. **Data Structures Used**:
   - The algorithm operates on an **array**.
   - It uses simple integer pointers (`low`, `mid`, `high`) and an index variable (`floorIndex`) to maintain the candidate.

---

## 4. Explanations + Code

### Detailed Explanation
- **Initialization**:  
  Start with the entire range of the array and no candidate floor.
- **Mid Calculation & Comparison**:  
  The mid element is compared with *x*. If it is less than or equal to *x*, it becomes the current best candidate and the algorithm searches for a potentially better (larger) candidate in the right half.
- **Pointer Adjustments**:
  - When `array[mid]` ≤ *x*, the left side is not guaranteed to have a closer candidate, so we move right.
  - When `array[mid]` > *x*, the search moves left.
- **Time Complexity**:  
  Since the search space is halved each time, the algorithm runs in **O(log n)**.

### C++ Implementation

```cpp
#include <iostream>
#include <vector>
using namespace std;

int findFloor(const vector<int>& arr, int x) {
    int low = 0, high = arr.size() - 1;
    int floorIndex = -1; // Indicates no floor found initially

    while (low <= high) {
        int mid = low + (high - low) / 2;

        // If exact match is found, it's the floor
        if (arr[mid] == x)
            return arr[mid];

        // If arr[mid] is less than x, it is a candidate
        if (arr[mid] < x) {
            floorIndex = mid;
            low = mid + 1;
        } else {
            // arr[mid] is greater than x, so search in the left half
            high = mid - 1;
        }
    }
    
    // If no candidate was found, return -1 to indicate floor does not exist
    return (floorIndex != -1) ? arr[floorIndex] : -1;
}

int main() {
    vector<int> arr = {1, 2, 8, 10, 10, 12, 19};
    int x = 5;
    
    int floorValue = findFloor(arr, x);
    if(floorValue != -1)
        cout << "The floor of " << x << " is " << floorValue << "." << endl;
    else
        cout << "No floor exists for " << x << " in the array." << endl;
    
    return 0;
}
```

**Explanation**:
- **Initialization**:  
  Pointers `low` and `high` cover the entire array, with `floorIndex` initialized to -1.
- **Midpoint and Comparison**:  
  The code checks whether the mid element equals *x* or is less than *x* (updating the candidate accordingly).
- **Return Value**:  
  The function returns the element at `floorIndex` (if found) or `-1` if no floor exists.

## 5. Animated Visualization (Interactive Demo)

Below is a **Python** code snippet using `matplotlib` and `ipywidgets` to create an interactive visualization of the floor-finding process using binary search. The visualization shows the array and how the pointers and candidate floor are updated at each step.

```python
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider
import numpy as np

# Sorted array for visualization
arr = [1, 2, 8, 10, 10, 12, 19]
target = 5

def floor_search_steps(arr, target):
    """
    Records each step of the binary search to find the floor.
    Each step is a tuple (low, mid, high, candidate_index).
    """
    steps = []
    low, high = 0, len(arr) - 1
    candidate_index = -1
    
    while low <= high:
        mid = low + (high - low) // 2
        # Update candidate if current mid is less than target
        if arr[mid] <= target:
            candidate_index = mid
        steps.append((low, mid, high, candidate_index))
        
        if arr[mid] == target:
            break
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return steps

steps = floor_search_steps(arr, target)

def draw_floor_step(step_idx=0):
    low, mid, high, candidate_index = steps[step_idx]
    
    plt.figure(figsize=(8, 3))
    plt.title(f"Step {step_idx+1}: low = {low}, mid = {mid}, high = {high}\nCandidate Index = {candidate_index}")
    indices = np.arange(len(arr))
    values = np.array(arr)
    
    # Create a bar plot for the array elements
    bars = plt.bar(indices, values, color='lightblue', edgecolor='black')
    
    # Highlight pointers
    bars[low].set_color('green')
    bars[mid].set_color('red')
    bars[high].set_color('purple')
    
    # If a candidate exists, highlight it in orange
    if candidate_index != -1:
        bars[candidate_index].set_color('orange')
    
    # Annotate each bar with its value
    for i, v in enumerate(arr):
        plt.text(i, v + 0.5, str(v), ha='center', fontsize=12)
    
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.xticks(indices)
    plt.ylim(0, max(arr) + 5)
    plt.show()

interact(draw_floor_step, step_idx=IntSlider(min=0, max=len(steps)-1, step=1, value=0));
```

**Visualization Explanation**:
- **Step Recording**:  
  The `floor_search_steps` function records the values of `low`, `mid`, `high`, and the current candidate (if any) at each iteration.
- **Drawing Function**:  
  The `draw_floor_step` function uses a bar plot to show:
  - **Green**: the `low` pointer.
  - **Red**: the `mid` pointer.
  - **Purple**: the `high` pointer.
  - **Orange**: the current candidate floor.
- **Interactivity**:  
  Using `ipywidgets.interact`, you can move through the steps to see how the candidate floor is updated during the binary search.
