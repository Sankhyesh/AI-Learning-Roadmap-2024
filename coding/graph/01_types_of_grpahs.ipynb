```python
from IPython.display import Markdown, display
import networkx as nx
import matplotlib.pyplot as plt

# Helper to draw graphs
def draw_graph(G, pos=None, title="", directed=False, weight_label=False):
    plt.figure(figsize=(5,4))
    if directed:
        nx.draw(G, pos or nx.spring_layout(G), with_labels=True, arrows=True, node_size=800)
    else:
        nx.draw(G, pos or nx.spring_layout(G), with_labels=True, node_size=800)
    if weight_label:
        edge_labels = nx.get_edge_attributes(G, 'weight')
        nx.draw_networkx_edge_labels(G, pos or nx.spring_layout(G), edge_labels)
    plt.title(title)
    plt.axis('off')
    plt.show()

# 1. Directed Graph

display(Markdown(r"""
◦ **Directed Graphs.**  
▪ Edges have a direction.  
▪ Represented with an arrow from one node to another.  
▪ If an edge goes from A to B (A -> B), A is connected to B, but B is not necessarily connected to A.  
▪ Example: A->B, B->C, A->D, C->D.
"""))
G = nx.DiGraph([("A","B"),("B","C"),("A","D"),("C","D")])
pos = {"A":(0,1),"B":(1,1),"C":(1,0),"D":(0,0)}
draw_graph(G, pos, "Directed Graph", directed=True)

# 2. Undirected Graph

display(Markdown(r"""
◦ **Undirected Graphs.**  
▪ Edges have no direction.  
▪ Represented by a simple line between two nodes (A—B means bidirectional).  
▪ Example: A-B, B-C, C-D, A-D, A-C.
"""))
G = nx.Graph([("A","B"),("B","C"),("C","D"),("A","D"),("A","C")])
pos = nx.circular_layout(G)
draw_graph(G, pos, "Undirected Graph")

# 3. Weighted Graph

display(Markdown(r"""
◦ **Weighted Graphs.**  
▪ Edges have a weight associated with them (e.g., distance or duration).  
▪ Example: Delhi—Mumbai (10 hrs), Delhi—Jaipur (4 hrs), Jaipur—Agra (3 hrs), Agra—Mumbai (8 hrs).
"""))
G = nx.Graph()
edges = [("Delhi","Jaipur",4),("Delhi","Mumbai",10),("Jaipur","Agra",3),("Agra","Mumbai",8)]
G.add_weighted_edges_from(edges)
pos = nx.spring_layout(G)
draw_graph(G, pos, "Weighted Graph", weight_label=True)

# 4. Unweighted Graph

display(Markdown(r"""
◦ **Unweighted Graphs.**  
▪ Edges do not carry weights; all connections are treated equally.
"""))
G = nx.Graph([("A","B"),("B","C"),("C","A")])
pos = nx.circular_layout(G)
draw_graph(G, pos, "Unweighted Graph")

# 5. Complete Graph

display(Markdown(r"""
◦ **Complete Graph.**  
Every node is connected to every other node.
"""))
G = nx.complete_graph(["A","B","C","D"])
pos = nx.circular_layout(G)
draw_graph(G, pos, "Complete Graph")

# 6. Disconnected Graph

display(Markdown(r"""
◦ **Disconnected Graph.**  
Multiple components with no edges between them.
"""))
G = nx.Graph([("A","B"),("B","C"),("C","D"),("E","F")])
pos = {**nx.spring_layout(G.subgraph(["A","B","C","D"])), **nx.spring_layout(G.subgraph(["E","F"]), center=(2,-1))}
draw_graph(G, pos, "Disconnected Graph")

# 7. Cyclic Graph

display(Markdown(r"""
◦ **Cyclic Graph.**  
Contains at least one cycle (closed loop).
"""))
G = nx.Graph([("A","B"),("B","C"),("C","D"),("D","A")])
pos = nx.circular_layout(G)
draw_graph(G, pos, "Cyclic Graph")

# 8. Acyclic Graph

display(Markdown(r"""
◦ **Acyclic Graph.**  
Contains no cycles.
"""))
G = nx.DiGraph([("A","B"),("B","C"),("A","C")])
pos = {"A":(0,1),"B":(1,1),"C":(1,0)}
draw_graph(G, pos, "Acyclic Graph", directed=True)

# 9. Euler Graph

display(Markdown(r"""
◦ **Euler Graph.**  
All vertices have even degree.
"""))
G = nx.Graph([("A","B"),("B","C"),("C","D"),("D","A")])
pos = nx.circular_layout(G)
draw_graph(G, pos, "Euler Graph")

# 10. Hamiltonian Graph

display(Markdown(r"""
◦ **Hamiltonian Graph.**  
Contains a cycle visiting each node exactly once.
"""))
G = nx.Graph([("A","B"),("B","C"),("C","D"),("D","A")])
pos = nx.circular_layout(G)
draw_graph(G, pos, "Hamiltonian Graph")
```
